```python


# Load dataset
# Select only relevant columns
df = df[["verse_number", "verse_in_sanskrit", "translation_in_english", "meaning_in_english", "translation_in_hindi"]]

# Drop any missing values (optional)
df.dropna(inplace=True)

# Display first few row
df.head()
```

```python
df["text"] = (
    df["verse_number"] + "\n\n" +  # Verse reference
    "📖 Sanskrit:\n" + df["verse_in_sanskrit"] + "\n\n" +  # Sanskrit Verse
    "🗣 Transliteration:\n" + df["translation_in_english"] + "\n\n" +  # English translation
    "💡 Meaning:\n" + df["meaning_in_english"]  # Meaning in English
)

# Save cleaned dataset
df[["text"]].to_csv("cleaned_bhagavad_gita.csv", index=False)

# Show an example
df["text"].head()
```
